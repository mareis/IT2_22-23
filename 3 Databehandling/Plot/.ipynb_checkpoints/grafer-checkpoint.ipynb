```python
import matplotlib.pyplot as plt

plt.plot([1,2],[4,10], 'o-', label="$Trym(x)=6x-4$")
plt.plot([1,2], [6,8], label="$Birk(x)=2x+4$")
plt.title("Birk vs Trym ned tusten")
plt.xlabel("Tid i sekunder (s)")
plt.ylabel("fart i m/s")
plt.legend() 
plt.show()
```

```python
import math 
def livsglede(år):
    if år<20:
        return 100*.90**år
    
    else:
        return 75/(1+math.exp(-0.2*(år-28)))

    
livsløp = [x for x in range(0,62)]

livsglede_liste = []

for år in livsløp:
    livsglede_liste.append(livsglede(år))


plt.plot(livsløp, livsglede_liste)
#plt.hlines(95, 0, 65, color="red")
plt.show()
```

```python
import math 
def livsglede(år):
    if år<20:
        return 100*.90**år
    
    else:
        return 75/(1+math.exp(-0.2*(år-28)))

    
livsløp = [x for x in range(0,62)]

livsglede_liste = []

for år in livsløp:
    livsglede_liste.append(livsglede(år))


plt.style.use("bmh")
#plt.style.use("ggplot")

plt.plot(livsløp, livsglede_liste)
plt.ylim(0,100)

#plt.hlines(95, 0, 65, color="red")
plt.show()
```
