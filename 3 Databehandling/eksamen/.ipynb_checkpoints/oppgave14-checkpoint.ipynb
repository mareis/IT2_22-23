# Figur 1

**Jeg teller opp og samler stoppestedene med flest og færest "avganger"**

```python
flest = df.value_counts("start_station_name").head(3)
færrest = df.value_counts("start_station_name").tail(3)

ftss = pd.concat([flest, færrest])
ftss
```

**Jeg velger søylediagram for å representer dataene**

```python
import matplotlib.pyplot as plt

ftss.plot.barh(ftss)
plt.title("3 mest -og minst populære startdestinasjonene")
plt.xlabel("Antal sykkelavganger")
plt.ylabel("Startlokasjon")
plt.show()
```

## Figur 2

```python
import numpy as np

frekvens = df.value_counts("end_station_name")
frekvens
```

```python
varighet = df.groupby('end_station_name')['duration'].sum()
```

```python
gjennomsnitt = varighet/frekvens
gjennomsnitt
```

```python
gsort = gjennomsnitt.sort_values(ascending=False)
gsort
```

```python
gsort.head(3).plot.barh()
plt.xlabel("Gjennomsnitsvarighet")
plt.ylabel("Sluttstasjon")
plt.show()
```

```python
df["dag"] = pd.to_datetime(df["started_at"]).dt.weekday

dager = df.value_counts("dag").sort_index()
dager.plot()
plt.xticks([0, 1, 2, 3, 4, 5, 6], ["Mandag", "Tirsdag", "Onsdag", "Torsdag", "Fredag", "Lørdag", "Søndag"], rotation=20)
plt.show()
```
