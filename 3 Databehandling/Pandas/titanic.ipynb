```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.style.use(['ggplot'])
sns.set_palette("deep", 10)

sns.get_dataset_names()
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
df=df.dropna()
df
```

```python
antall = df['species'].value_counts()
antall
```

```python
sns.barplot(x=antall.index, y=antall)
plt.show()
```

```python
antall.plot.bar()
plt.show()
```

```python
antall = df.groupby('species').count()['island']
sns.barplot(x=antall.index, y=antall)
plt.show()
```
