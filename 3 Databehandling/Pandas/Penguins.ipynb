```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.style.use(['ggplot'])
sns.set_palette("deep", 10)
df = sns.load_dataset('penguins')
df=df.dropna()

df
```

```python
df.species.value_counts()

#På øyene finner man 146 Adelie pinguiner
#På øyene finner man 119 Gentoo pinguiner
#På øyene finner man 68 Chrinstrap pinguiner
```

```python
df.groupby('island')['body_mass_g'].sum()

#kroppsvekt for alle penguiner på øya Biscoe er 769225
#kroppsvekt for alle penguiner på øya Dream er 457425
#kroppsvekt for alle penguiner på øya Torgersen er 174300
```

```python
dfmax = df.groupby('species')['flipper_length_mm'].max()
dfmin = df.groupby('species')['flipper_length_mm'].min()

print(dfmax)
print(dfmin)

#Minste flipper lengde for hver art: Adelie 172.0, Chinstrap 178.0, Gentoo 203.0
#Største flipper lengde for hver art: Adelie 210.0, Chinstrap 212.0, Gentoo 231.0
```

```python
df.groupby('sex')['body_mass_g'].median()

#Median kroppsvekt for damer er 3650
#Median kroppsvekt for menn er 4300
```
