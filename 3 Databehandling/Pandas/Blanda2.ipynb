```python
import pandas as pd
```

```python
df = pd.read_csv("Lonn_2015-2020.csv", "\t")
df.tail()
```

```python
df.plot(x="aar", y="Gjennomsnitt Maanedslnn (kr)")
```

```python
import matplotlib.pyplot as plt

menn = df.loc[df['kjonn'] == "Menn", 'Gjennomsnitt Maanedslnn (kr)']
kvinner = df.loc[df['kjonn'] == "Kvinner", 'Gjennomsnitt Maanedslnn (kr)']
aar = df.loc[df['kjonn'] == "Menn", 'aar']


plt.plot(aar, menn)
plt.plot(aar, kvinner)
```

```python
import seaborn as sns

sns.catplot(x = "aar",                                # x variable name
            y = "Gjennomsnitt Maanedslnn (kr)",       # y variable name
            hue = "kjonn",                            # group variable name
            data = df,                                # dataframe to plot
            kind = "bar")
```

```python
sns.relplot(
    data=df, kind="line",
    x="aar", y="Gjennomsnitt Maanedslnn (kr)",
    hue="kjonn", 
)
```
