```python
import pandas as pd

df = pd.read_csv("fritidsboliger_moh_2019.csv", sep=";",index_col=0, header=None).T
df = df.astype({"Meter over havet": str, "Fritidsbygg": int})

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_short = df[:14]

sns.catplot(y = "Meter over havet",                                # x variable name
            x = "Fritidsbygg",       # y variable name
            data = df_short[::-1],                                # dataframe to plot
            kind = "bar").set(title='Fritidboliger')

plt.show()
```
